```python
# Drop rows that have missing Customer ID (because you can’t tie to a customer)
df=df.dropna(subset=['CustomerID'])
df
```

```python
# Fill missing Description or other columns using a placeholder (e.g. "Unknown")
df['Description']=df['Description'].fillna('Unkknown')
df
```

**Date Parsing / Format Correction**

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce', format='mixed')
df
```

```python
print(df['InvoiceDate'].isnull().sum())
```

**Detect & Fix Bad or Outlier Data**

```python
# Some transactions might have negative quantity (returns) — detect those and decide whether to treat as returns or drop
neg_quat=df[df['Quantity']<0]
print(neg_quat)
```

```python
# Some unit prices might be zero or negative — fix/clean these (e.g. drop or adjust)
df = df[df['Quantity'] >= 0]
```

```python
df['IsReturn'] = df['Quantity'] < 0
```

```python
bad_prices = df[df['UnitPrice'] <= 0]
print("Zero or Negative Prices:\n")
bad_prices
```

```python
# Replace zero or negative prices with the mean of valid prices
mean_price = df[df['UnitPrice'] > 0]['UnitPrice'].mean()
df.loc[df['UnitPrice'] <= 0, 'UnitPrice'] = mean_price
df
```

```python
print("Data after cleaning:")
print(df.describe())
```

**Duplicates**

```python
duplicates = df.duplicated(subset=['InvoiceNo', 'StockCode'], keep=False)
print("Number of duplicated invoice-item entries:", duplicates.sum())
```

```python
df = df.drop_duplicates(subset=['InvoiceNo', 'StockCode'], keep='first')
```

```python
duplicates_after = df.duplicated(subset=['InvoiceNo', 'StockCode']).sum()
print("Number of duplicates after cleaning:", duplicates_after)
```

**Filtering & Sorting**

```python
# Filter transactions where Quantity > 1000 (suspicious outliers
suspicious_transactions=df[df['Quantity']>1000]
suspicious_transactions
```

```python
# Sort by UnitPrice * Quantity descending (i.e. highest revenue orders)
df['TotalRevenue']=df['UnitPrice']*df['Quantity']
df=df.sort_values(by='TotalRevenue',ascending=False)
df
```

**Merge / Enrich Data**

```python
# Create a small custom DataFrame product_info with columns StockCode, Category, Supplier
product_info=pd.DataFrame({
                  'StockCode':['a','b','c'],
                  'Category':['homeapp','ele','plas'],
                  'Supplier':['abc','def','jkl']
})
print(product_info)
```

```python
# Convert its date (if needed) and merge with the main dataset on StockCode (left join)
df = df.merge(product_info, on='StockCode', how='left')
df.head()
```

```python
df = df.drop(columns=['Category_x', 'Supplier_x', 'Category_y', 'Supplier_y'])
```

```python
print(df.columns)
```

**Group & Aggregation**

```python
# Create a new column for Total Revenue
df['TotalRevenue'] = df['Quantity'] * df['UnitPrice']

# Group by Country and compute total revenue, mean revenue per order, and count of orders
country_revenue_stats = df.groupby('Country').agg(
    TotalRevenue=('TotalRevenue', 'sum'),
    MeanRevenuePerOrder=('TotalRevenue', 'mean'),
    OrderCount=('InvoiceNo', 'nunique')
).reset_index()

print("Revenue statistics by Country:")
print(country_revenue_stats)

# Use pd.cut() to bucket UnitPrice and then group by that price range
df['PriceRange'] = pd.cut(df['UnitPrice'], bins=[0, 1, 5, 10, 100, df['UnitPrice'].max()],
                          labels=['Cheap', 'Medium', 'Expensive', 'Very Expensive', 'Luxury'])

price_range_stats = df.groupby('PriceRange', observed=False).agg(
    PurchaseCount=('InvoiceNo', 'count'),
    AverageRevenue=('TotalRevenue', 'mean')
)

print("\nPurchase and Revenue statistics by Price Range:")
print(price_range_stats)
```

***`**Correlation / Relationships**`***

```python
# Calculate the correlation matrix for the numeric columns
correlation_matrix = df[['Quantity', 'UnitPrice', 'TotalRevenue']].corr()

print("Correlation Matrix:")
print(correlation_matrix)

# Interpret the correlations
print("\nInterpretation of Correlation Matrix:")
if correlation_matrix.loc['Quantity', 'TotalRevenue'] > 0.5:
    print("There is a strong positive correlation between Quantity and TotalRevenue.")
elif correlation_matrix.loc['Quantity', 'TotalRevenue'] < -0.5:
    print("There is a strong negative correlation between Quantity and TotalRevenue.")
else:
    print("There is a weak or no strong correlation between Quantity and TotalRevenue.")

if correlation_matrix.loc['UnitPrice', 'TotalRevenue'] > 0.5:
    print("There is a strong positive correlation between UnitPrice and TotalRevenue.")
elif correlation_matrix.loc['UnitPrice', 'TotalRevenue'] < -0.5:
    print("There is a strong negative correlation between UnitPrice and TotalRevenue.")
else:
    print("There is a weak or no strong correlation between UnitPrice and TotalRevenue.")

if correlation_matrix.loc['Quantity', 'UnitPrice'] > 0.5:
    print("There is a strong positive correlation between Quantity and UnitPrice.")
elif correlation_matrix.loc['Quantity', 'UnitPrice'] < -0.5:
    print("There is a strong negative correlation between Quantity and UnitPrice.")
else:
    print("There is a weak or no strong correlation between Quantity and UnitPrice.")
```
