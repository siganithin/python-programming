```python
farm_a=np.array([[10,12,11],[9,11,10],[12,13,14],[8,7,9]])
farm_b=np.array([[20,19,21],[18,20,19],[22,23,21],[17,18,16]])
farm_c=np.array([[15,16,14],[14,13,15],[16,17,18],[12,11,13]])
print(farm_a)
print(farm_b)
print(farm_c)
```

```python
production_all=np.stack([farm_a,farm_b,farm_c],axis=0)
print(production_all)
```

```python
print("Shape:",production_all.shape)
print("size:",production_all.size)
print("data type:",production_all.dtype)
print("number of dimension:",production_all.ndim)
```

```python
total_per_farm=production_all.sum(axis=(1,2))
print(total_per_farm)
```

Compute mean, std, and var per farm.

```python
mean_per_farm=production_all.mean(axis=(1,2))
std_per_farm=production_all.std(axis=(1,2))
var_per_farm=production_all.var(axis=(1,2))
print(mean_per_farm)
print(std_per_farm)
print(var_per_farm)
```

Per-hour matrix multiplication idea (matrix op)

```python
hour_weights=np.array([0.5, 1.0, 0.75])
weighted_per_panel = np.dot(production_all, hour_weights)
print(weighted_per_panel)
```

Broadcasting adjustments

```python
offset=0.5
production_all = production_all.astype(float)
production_all[1] += offset
print(production_all)
```

Element-wise arithmetic (ufuncs: add/sub/mul/div/power/mod)
Compute a normalized version where you:
subtract the per-farm mean (use broadcasting),
divide by the per-farm std (broadcasting) — i.e., z-score normalize each farm's panel-hour readings.

```python
mean_per_farm = production_all.mean(axis=(1,2))
std_per_farm = production_all.std(axis=(1,2))
print(mean_per_farm)
print(std_per_farm)
```

```python
mean_per_farm = mean_per_farm[:, np.newaxis, np.newaxis]  # shape (3,1,1)
std_per_farm = std_per_farm[:, np.newaxis, np.newaxis]    # shape (3,1,1)
print(mean_per_farm)
print(std_per_farm)
```

```python
normalized_production = (production_all - mean_per_farm) / std_per_farm
print(normalized_production)
```
